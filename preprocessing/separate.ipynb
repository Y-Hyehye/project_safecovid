```python
import pandas as pd
```

```python
# 안심식당 데이터 불러오기(df 정의)
df=pd.read_csv('res.csv', encoding='cp949')
```

```python
df
```

```python
# df 전처리
# 주소 하나로 합치기
df['주소']=df['주소1'] + " " + df['주소2']
```

```python
# 필요한 데이터 추출
df=df[['경도','사업장명','업종상세','위도','전화번호', '주소', '시도명']]
```

```python
df.dropna(subset=['위도', '경도'], inplace=True)
```

```python
# 시도명에 따라 10개의 지역으로 각 df 나누기
df_seoul=df[df['시도명'].str.contains('서울', na=False)]
df_gyeounggi=df[df['시도명'].str.contains('경기|인천', na=False)]
df_chungnam=df[df['시도명'].str.contains('충청남도|대전|세종', na=False)]
df_chungbuk=df[df['시도명'].str.contains('충청북도', na=False)]
df_jeonnam=df[df['시도명'].str.contains('전라남도|광주', na=False)]
df_gangwon=df[df['시도명'].str.contains('강원', na=False)]
df_gyeongbuk=df[df['시도명'].str.contains('경상북도|대구', na=False)]
df_gyeongnam=df[df['시도명'].str.contains('경상남도|부산|울산', na=False)]
df_jeju=df[df['시도명'].str.contains('제주', na=False)]
df_jeonbuk=df[df['시도명'].str.contains('전라북도', na=False)]
```

```python
# 시도명 데이터(열) 삭제
del df_seoul['시도명']
del df_gyeounggi['시도명']
del df_chungnam['시도명']
del df_chungbuk['시도명']
del df_jeonnam['시도명']
del df_gangwon['시도명']
del df_gyeongbuk['시도명']
del df_gyeongnam['시도명']
del df_jeju['시도명']
del df_jeonbuk['시도명']
```
